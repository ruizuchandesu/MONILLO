```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"C:\Users\sidne\OneDrive\Desktop\datasets\current-power.csv",
                 delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x=df['Current']
y=df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power')
plt.ylabel('Power')
plt.xlabel('Current')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color="#164085", alpha=0.6, label='Data points')

plt.plot(x, line, color="#0C5FDB", label=f"Regression Line (r = {r_value:.3f})")

plt.title("Device Current vs Power Response")
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

## Air Quality

```python
air = pd.read_csv(r"C:\Users\sidne\OneDrive\Desktop\datasets\air-quality.csv",
                  delimiter=",")
air
```

## NOx

```python
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color="#AD8223", alpha=0.6, label='Data points')

plt.plot(x, line, color="#1166E5", label=f"Regression Line (r = {r_value:.3f})")

plt.title("NOx vs Humidity Corrrelation")
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```

## NO2

```python
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(7,4))
plt.scatter(x, y, color="#D520C0", alpha=0.6, label='Data points')

plt.plot(x, line, color="#800606", label=f"Regression Line (r = {r_value:.3f})")

plt.title("NO2 vs Humidity Corrrelation")
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (ug/cu.m)')
plt.legend()
plt.show()
```
