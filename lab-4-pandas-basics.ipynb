# Descriptive Statistics

```python
resistance = np.array([
    [10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
    [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]
]).T
resistance
```

```python
df = pd.DataFrame(resistance, columns = ['Jaguar', 'Panther'])
df
```

```python
type(df)
```

```python
df['Panther']
```

```python
df.info()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df.fillna(100)
```

```python
df['Panther'].fillna(df['Panther'].mean())
```

```python
df
```

```python
df.describe()
```

```python
# Median
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar Median = {jag_median:.2f} ohms')
print(f'Panther Median = {pan_median:.2f} ohms')
```

```python
# Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
jag_mode
```

```python
type(jag_mode)
```

```python
print('Jaguar Mode:')

for i in jag_mode:
    print(i)
```

```python
print('Panther Mode:')

for i in pan_mode:
    print(i)
```

```python
# Jaguar IQR
Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar IQR = {Q3 - Q1:.2f} ohms')
```

```python
# Panther IQR
Q1 = df['Panther'].quantile(0.25)
Q3 = df['Panther'].quantile(0.75)

print(f'Panther IQR = {Q3 - Q1:.2f} ohms')
```

# Bar Graph

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure()
bars = plt.bar(['Jaguar','Panther'],[jag_mean,pan_mean],
        color="#326cb8ff")

plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')

#label
plt.bar_label(bars,padding=3)
plt.margins(y=0.5)
plt.show()
```
