```python
#imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Z-score

```python
# creating a dataframe
df = pd.DataFrame([1, 2, 2, 3, 3, 3, 4, 4, 5],
                  columns=['Dataset'])
df
```

```python
# summary of statistics
df.describe()
```

```python
# Z-scores
df['Z-score'] = stats.zscore(df['Dataset'],ddof=1)
df
```

```python
df['Z-score'].mean()
```

```python
df['Z-score'].std(ddof=1)
```

# Histogram

```python
# histogram plot
plt.hist(df['Current'], bins = 5,
         density=True,
         alpha = 0.8,
         color = 'Brown',
         edgecolor = 'Black')

plt.title('Histogram of Current Values',
          font = 'Palatino Linotype',
          fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

# Normal Distribution

```python
# normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = 'Brown',
         linewidth=2.9)

plt.title('Normal Distribution of Current Values',
         font = 'Palatino Linotype',
         fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
# Histogram
plt.hist(df['Current'], bins = 5,
        density = True,
        alpha = 0.8,
        color = 'Brown',
        edgecolor = 'Black')

plt.title('Histogram of Current Values',
         font = 'Palatino Linotype',
         fontsize = 12)

# Normal Distribution plot
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
         color = 'Black',
         linewidth=2.9)

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

# Standard Normal Distribution

```python
# x-axis
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()
x = np.linspace(x_min, x_max, 100)

p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
         color = 'Brown',
         linewidth=2.9)

plt.title('Standard Normal Distribution of Current Values N(0,1)',
         font = 'Palatino Linotype',
         fontsize = 12)

plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```
